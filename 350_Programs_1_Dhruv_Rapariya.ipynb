# 50. Check Whether number is armstrong or not

```python
num = int(input("Enter a number: "))

original = num

n = len(str(num))

sum = 0

while num > 0:
    digit = num % 10        
    sum += digit ** n       
    num = num // 10         

if sum == original:
    print("It is an Armstrong number")
else:
    print("It is NOT an Armstrong number")
```

# 92. Given a list of strings, create a new list containing only strings that are anagrams of the first string.

```python
def find_anagrams(words):
    first = words[0]
    sorted_first = sorted(first)
    
    result = []
    
    for word in words:
        if sorted(word) == sorted_first:
            result.append(word)
    
    return result


# Example
words = ["listen", "silent", "enlist", "hello", "inlets"]
print(find_anagrams(words))
```

```python
queue = []

def enqueue():
    item = input("Enter element to enqueue: ")
    queue.append(item)
    print(f"{item} added to queue")

def dequeue():
    if not queue:
        print("Queue is empty")
    else:
        removed = queue.pop()
        print(f"{removed} removed from queue")

def display():
    print("Queue:", queue)

while True:
    print("\n1. Enqueue")
    print("2. Dequeue")
    print("3. Display")
    print("4. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        enqueue()
    elif choice == 2:
        dequeue()
    elif choice == 3:
        display()
    elif choice == 4:
        break
    else:
        print("Invalid choice")
```

# 98. Given a list of numbers, generate all possible pairs (a, b) where a < b and sum of pair is even.

```python
nums = [1, 2, 3, 4, 5]

result = []

for i in range(len(nums)):
    for j in range(i + 1, len(nums)):
        a, b = nums[i], nums[j]
        
        if (a + b) % 2 == 0 and a < b:
            result.append((a, b))

print(result)
```

# 99. Given a list of numbers, generate all possible pairs (a, b) where a < b and sum of pair is even. using list comprehension

```python
nums = [1, 2, 3, 4, 5]

result = [(a, b) for i, a in enumerate(nums)
                   for b in nums[i+1:]
                   if (a + b) % 2 == 0]

print(result)
```

# 139. Generator Example – Fibonacci Sequence using yield

```python
def fibonacci(n):
    a, b = 0, 1
    count = 0
    while count < n:
        yield a
        a, b = b, a + b
        count += 1

for num in fibonacci(5):
    print(num)
```

# 158. write a lambda function that checks if a number is prime and use it with filter on list 1 - 20

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

nums = range(1, 21)
primes = list(filter(is_prime, nums))

print(primes)
```
